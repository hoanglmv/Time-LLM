```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu
# Sử dụng raw string (r'...') để tránh lỗi ký tự đặc biệt trong đường dẫn Windows
file_path = r'E:\Project\Time-LLM\dataset\electricity\electricity.csv'
df = pd.read_csv(file_path)

# Chuyển đổi cột date sang datetime và đặt làm index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# 2. EDA dữ liệu
print("Kích thước dữ liệu (Shape):", df.shape)
print("\nThông tin dữ liệu (Info):")
print(df.info())
print("\nThống kê mô tả (Describe):")
display(df.describe()) # Nếu chạy trong Jupyter, dùng display() sẽ đẹp hơn print()

# 3. Vẽ hình minh họa dữ liệu trong 2 tuần
# Lấy 2 tuần dữ liệu đầu tiên (14 ngày * 24 giờ = 336 dòng)
df_2weeks = df.iloc[:336]

plt.figure(figsize=(15, 6))

# Vẽ 5 cột đầu tiên để minh họa xu hướng (tránh vẽ tất cả 321 cột sẽ rất rối)
cols_to_plot = df.columns[:5] 
for col in cols_to_plot:
    plt.plot(df_2weeks.index, df_2weeks[col], label=f'Feature {col}')

plt.title('Biểu đồ tiêu thụ điện trong 2 tuần đầu (5 features đầu tiên)')
plt.xlabel('Thời gian')
plt.ylabel('Giá trị tiêu thụ')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
